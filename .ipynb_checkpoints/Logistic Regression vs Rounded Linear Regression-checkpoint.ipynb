```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.model_selection import train_test_split, cross_val_predict
from sklearn.metrics import confusion_matrix, precision_score, recall_score
```

## Logistic Regression - Red Wine

```python
LogModel_R = LogisticRegression(C=10**10, max_iter=50000)
```

```python
df = pd.read_csv("winequality-red.csv", sep=';')
X_R = df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']].values
y_R = df['quality'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_R, y_R, test_size=0.25, random_state=55)
```

```python
LogModel_R.fit(X_train, y_train)
```

```python
y_pred = LogModel_R.predict(X_test)
```

```python
plt.scatter(y_test, y_pred)
plt.title('Logistic Regression - Red Wine')
plt.xlabel('Real Values')
plt.ylabel('Prediction')
plt.grid()
plt.show()
```

```python
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True)
plt.xlabel('Predictions')
plt.ylabel('Real Values')
```

```python
print('Precision: ', precision_score(y_test, y_pred, average='weighted'))
print('Recall: ', recall_score(y_test, y_pred, average='weighted'))
#Sale una advertencia debido a que algunas columnas no tienen datos para comparar.
```

## Linear Regression (Rounded) - Red Wine

```python
LinModel_R = LinearRegression()
```

```python
LinModel_R.fit(X_train, y_train)
```

```python
y_predLR = LinModel_R.predict(X_test).round()
```

```python
plt.scatter(y_predLR, y_test)
plt.ylabel('Real Values')
plt.xlabel('Prediction')
plt.grid()
plt.show()
```

```python
cm = confusion_matrix(y_test, y_predLR)
plt.figure(figsize=(8,8))
sns.heatmap(cm, annot=True)
plt.xlabel('Predictions')
plt.ylabel('Real Values')
```

```python
print('Precision: ', precision_score(y_test, y_predLR, average='weighted'))
print('Recall: ', recall_score(y_test, y_predLR, average='weighted'))
#Sale una advertencia debido a que algunas columnas no tienen datos para comparar.
```
